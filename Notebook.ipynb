### Data Preprocessing

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
data1 = data.copy()
```

```python
data1.reset_index(inplace = True)
```

```python
data1
```

```python
data1['title']
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []

for i in range(len(data1)):
    title = re.sub('[^a-zA-Z]', ' ', data1['title'][i])
    title = title.lower()
    title = title.split()
    
    title = [ps.stem(word) for word in title if not word in stopwords.words('english')]
    title = " ".join(title)
    corpus.append(title)
    
    if len(corpus) % 1000 == 0:
        print("Corpus size =", len(corpus))
```

```python
len(corpus)
```

```python
corpus
```

```python
cv = CountVectorizer(max_features = 5000, ngram_range = (1, 3))
X = cv.fit_transform(corpus)
```

```python
X = X.toarray()
```

```python
X
```

```python
X.shape
```

```python
Y = data1.label
```

```python
Y = Y.to_numpy()
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, random_state = 10)
```

```python
x_test, x_cv, y_test, y_cv = train_test_split(x_test, y_test, test_size = 0.5, random_state = 10)
```

```python
np.save('x_train', x_train)
np.save('y_train', y_train)
np.save('x_test', x_test)
np.save('y_test', y_test)
np.save('x_cv', x_cv)
np.save('y_cv', y_cv)
```

### Shallow Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(x_train, y_train)
```

```python
train_score = mnb.score(x_train, y_train)

print("Training score = {:.2f}".format(train_score))
```

```python
cv_score = mnb.score(x_cv, y_cv)

print("validation score = {:.2f}".format(cv_score))
```

```python
test_score = mnb.score(x_test, y_test)

print("Testing score = {:.2f}".format(test_score))
```

```python
pred = mnb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, pred)
cm_dis = ConfusionMatrixDisplay(cm)
```

```python
cm_dis.plot()
```
